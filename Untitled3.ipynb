```python
# Installer PyTorch si nécessaire (déjà installé sur Colab)
# !pip install torch matplotlib

import torch
import torch.nn as nn
import torch.autograd as autograd
import matplotlib.pyplot as plt

# -------------------------------
# Définition du réseau PINN
# -------------------------------
class PINN(nn.Module):
    def __init__(self, layers):
        super(PINN, self).__init__()
        self.layers = nn.ModuleList()
        for i in range(len(layers)-1):
            self.layers.append(nn.Linear(layers[i], layers[i+1]))
        self.activation = nn.Tanh()

    def forward(self, x):
        for i in range(len(self.layers)-1):
            x = self.activation(self.layers[i](x))
        x = self.layers[-1](x)
        return x

# -------------------------------
# Paramètres physiques
# -------------------------------
D = 1e-9   # Diffusion [m^2/s]
v = 1e-3   # Vitesse du fluide [m/s]
r_max = 1e-4  # Vitesse maximale réaction [mol/(m^3 s)]
K_S = 0.1     # Constante Monod [mol/m^3]

# -------------------------------
# Création du réseau
# -------------------------------
layers = [2, 50, 50, 1]  # entrée: (x,t), sortie: C_A
model = PINN(layers)

# Optimiseur
optimizer = torch.optim.Adam(model.parameters(), lr=1e-3)

# -------------------------------
# Points d'entraînement (collocation points)
# -------------------------------
N_points = 100
x = torch.linspace(0,1,N_points)
t = torch.linspace(0,1,N_points)
X, T = torch.meshgrid(x, t, indexing='ij')
X_train = torch.cat([X.reshape(-1,1), T.reshape(-1,1)], dim=1)
X_train.requires_grad = True

# -------------------------------
# Entraînement du PINN
# -------------------------------
epochs = 1000
for epoch in range(epochs):
    optimizer.zero_grad()

    C = model(X_train)

    # Dérivées pour le PINN
    C_t = autograd.grad(C, X_train, grad_outputs=torch.ones_like(C), create_graph=True)[0][:,1:2]
    C_x = autograd.grad(C, X_train, grad_outputs=torch.ones_like(C), create_graph=True)[0][:,0:1]
    C_xx = autograd.grad(C_x, X_train, grad_outputs=torch.ones_like(C_x), create_graph=True)[0][:,0:1]

    # Réaction Monod
    r = r_max * C / (K_S + C)

    # Résidu de l'équation de bilan de masse
    residual = C_t + v*C_x - D*C_xx - r
    loss = torch.mean(residual**2)

    loss.backward()
    optimizer.step()

    if epoch % 100 == 0:
        print(f"Epoch {epoch}, Loss = {loss.item():.6e}")

# -------------------------------
# Visualisation de C_A(x,t)
# -------------------------------
C_pred = model(X_train).detach().numpy()
C_pred = C_pred.reshape(N_points, N_points)

plt.figure(figsize=(8,5))
plt.pcolormesh(T.detach().numpy(), X.detach().numpy(), C_pred, shading='auto')
plt.colorbar(label='Concentration C_A [mol/m^3]')
plt.xlabel('Temps t [s]')
plt.ylabel('Position x [m]')
plt.title('Simulation PINN : Concentration dans le bioréacteur')
plt.show()
```
